```python
import matplotlib.pyplot as plt
import pickle
import numpy as np

path = '/data0/zuchuan/mae_output/cls.drop{}/fine_tune_accuracy.pickle'
for i in np.arange(0.1, 0.9, 0.1):
    name = path.format(int(i*10) / 10)
    print(name)
    with open(name, 'rb') as h:
        rs = pickle.load(h)
    rs = np.array(rs)
    fig, axe = plt.subplots(nrows=1, ncols=1)
    axe.plot(rs[:, 0], rs[:, 2])
    axe.set_title(str(int(i * 10) / 10) + " / "+ str(np.max(rs[:, 2])))
    axe.set_ylim([0, 0.4])
    axe.grid()
```
